```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ramp_signal (start, end, step=1 ):
    return np.arange(start, end, step)
```

```python
plt.plot(ramp_signal(-5, 10, 1))

plt.title('Ramp Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
