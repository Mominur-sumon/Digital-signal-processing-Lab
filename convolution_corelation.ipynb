```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Convolution 
def convolution(x, h):
    len_x = len(x)
    len_h = len(h)
    len_y = len_x + len_h - 1
    y = [0] * len_y 

    for i in range(len_y):
        for j in range(len_h):
            if i - j >= 0 and i - j < len_x:
                y[i] += x[i - j] * h[j]
    
    return y
```

```python
x = [1,2,3,4,5]
h = [2,3,4]

y_convolution = convolution(x, h)
y_correlation = convolution(x, h[::-1])
print(y_convolution)
print(y_correlation)


```

```python
# Ploting the results

plt.figure(figsize=(16, 8))
plt.subplot(2,3,1)
plt.title(f'x[n]')
plt.stem(x)

plt.subplot(2,3,2)
plt.title(f'h[n]')
plt.stem(h)

plt.subplot(2,3,3)
plt.title(f'x[n] * h[n]')
plt.stem(y_convolution)

plt.subplot(2,3,4)
plt.title(f'x[n]')
plt.stem(x)

plt.subplot(2,3,5)
plt.title(f'h[n]')
plt.stem(h)

plt.subplot(2,3,6)
plt.title(f'x[n] . h[n]')
plt.stem(y_correlation)
```
