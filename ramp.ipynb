```python
import numpy as np
import matplotlib.pyplot as plt

def ramp(t, slope=1):
    return np.where(t >= 0, slope * t, 0)

# Generate time values
t = np.linspace(-5, 5, 1000)

# Generate ramp signal
ramp_signal = ramp(t)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(t, ramp_signal, label='Ramp')
plt.title('Ramp Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```
