```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
sampling_rate = 8000
N = 8 
ts = 1 / sampling_rate 
n = np.arange(N)
x = np.sin(2 * np.pi * 1000 * n * ts) + 0.5 * np.sin(2 * np.pi * 2000 * n * ts + 3 * np.pi / 4)
```

```python
plt.plot(n, x)
plt.xlabel('Sample number n')
plt.ylabel('Amplitude')
plt.title('Signal x(n)')
plt.grid(True)
plt.show()
```

```python
dft_s = np.zeros(N, dtype=complex)
for n in range(N):
    for m in range(N):
        dft_s[n] += x[m] * np.exp(-2j * np.pi * n * m / N)
    dft_s[n] = np.round(dft_s[n], 2)
```

```python
magnitude = np.abs(dft_s)
plt.stem(range(N), magnitude)
plt.xlabel('Frequency bin m')
plt.ylabel('Magnitude')
plt.title('DFT Magnitude')
plt.grid(True)
plt.show()
```

```python
phase = np.arctan(dft_s) * 180 / np.pi
plt.stem(range(N), phase)
plt.xlabel('Frequency bin m')
plt.ylabel('Phase (degrees)')
plt.title('DFT Phase')
plt.grid(True)
plt.show()
```

```python
plt.stem(range(N), dft_s.real)
plt.xlabel('Frequency bin m')
plt.ylabel('Real part')
plt.title('DFT Real Part')
plt.grid(True)
plt.show()
```

```python
plt.stem(range(N), dft_s.imag)
plt.xlabel('Frequency bin m')
plt.ylabel('Imaginary part')
plt.title('DFT Imaginary Part')
plt.grid(True)
plt.show()
```

```python
idft_s = np.zeros(N, dtype=complex)
for n in range(N):
    for m in range(N):
        idft_s[n] += dft_s[m] * np.exp(2j * np.pi * n * m / N)
    idft_s[n] = idft_s[n] / N
```

```python
plt.plot(range(N), [x.real for x in idft_s])
plt.xlabel('Sample number n')
plt.ylabel('Amplitude')
plt.title('IDFT Real Part')
plt.grid(True)
plt.show()
```
