```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Variables 
start = 0 
stop = 0.01
phase =90 
frequency = 500 
amplitude = 5

# Conversion of phase to radians
phase_radian = np.radians(phase)
```

```python
# Given Function:
# y = 5 sin(2 _ pi _ 500t + 90)
```

```python
# Base Signal 
sampling_rate = 4000
samples_count = sampling_rate*(stop-start)

# Generate the Signal 
t_base = np.linspace(start=start, stop = stop , num=int(samples_count))
y_base = amplitude * np.sin(2*np.pi*frequency*t_base + phase_radian)

# Plot the Signal
plt.figure(figsize=(10,6))
# plt.plot(t_base, y_base)
plt.stem(t_base, y_base)
plt.title("Base Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

```python
# Upsampling the Signal
upsampling_rate = 8000
upsampled_count = upsampling_rate*(stop-start)

# Conversion of Base Signal to Upsampled Signal
phase_radian = np.radians(phase)

# Generate the Signal
t_upsampled = np.linspace(start=start, stop = stop , num=int(upsampled_count))
y_upsampled = amplitude * np.sin(2*np.pi*frequency*t_upsampled + phase_radian)

plt.figure(figsize=(10,6))
plt.stem(t_upsampled, y_upsampled)
plt.title("Upsampled Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

```python
# Downsampling the Signal
downsampling_rate = 1000
downsampled_count = downsampling_rate*(stop-start)

# Conversion of Base Signal to Downsampled Signal
phase_radian = np.radians(phase)

# Generate the Signal

t_downsampled = np.linspace(start=start, stop = stop , num=int(downsampled_count))
y_downsampled = amplitude * np.sin(2*np.pi*frequency*t_downsampled + phase_radian)

plt.figure(figsize=(10,6))
plt.stem(t_downsampled, y_downsampled)
plt.title("Downsampled Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

```python
# Generate Signal 
plt.figure(figsize=(15,4))
plt.subplot(1,3,1)
plt.stem(t_base, y_base)
plt.title("Base Signal")

plt.subplot(1,3,2)
plt.stem(t_upsampled, y_upsampled)
plt.title("Upsampled Signal")


plt.subplot(1,3,3)
plt.stem(t_downsampled, y_downsampled)
plt.title("Downsampled Signal")
plt.show()
```
