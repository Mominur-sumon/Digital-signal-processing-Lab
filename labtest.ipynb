```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generating Orginal Signal
N = 512 
t = np.arange(0,0.01, 0.01/N)
x_t = 10*np.sin(2*np.pi *40*t) + 20*np.sin(2*np.pi*80*t) + 40*np.sin(2*np.pi*160*t) 
print("The number of sample is ", len(x_t))
```

```python
n = np.arange(512)
plt.plot(n,x_t)
plt.title("x_t = 10*np.sin(2*np.pi *40*t) + 20*np.sin(2*np.pi*80*t) + 40*np.sin(2*np.pi*160*t) ")
plt.xlabel("Time(S)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
# Sampling the signal with 400 HZ
Fs = 400
Ts = 1/Fs
n = np.arange(0,0.1, Ts)
x_n = 10*np.sin(2*np.pi *40*n) + 20*np.sin(2*np.pi*80*n) + 40*np.sin(2*np.pi*160*n)
print("The number of sample is ", len(x_n))
plt.stem(n,x_n)
plt.title("x_n = 10*np.sin(2*np.pi *40*n) + 20*np.sin(2*np.pi*80*n) + 40*np.sin(2*np.pi*160*n) ")
plt.xlabel("Time(S)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python

# Oversample the signal
Oversample_rate = 5
x_n_oversampled = 10*np.sin(2*np.pi *40*n*Oversample_rate) + 20*np.sin(2*np.pi*80*n*Oversample_rate) + 40*np.sin(2*np.pi*160*n*Oversample_rate)
print("The number of sample is ", len(x_n_oversampled))
```

```python
plt.stem(n,x_n_oversampled )
plt.title("Over sampled by 5")
plt.xlabel("Time(S)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
# Under sample the signal   
Under_sample_rate = 2
x_n_undersampled = 10*np.sin(2*np.pi *40*n/Under_sample_rate) + 20*np.sin(2*np.pi*80*n/Under_sample_rate) + 40*np.sin(2*np.pi*160*n/Under_sample_rate)
print("The number of sample is ", len(x_n_undersampled))
```

```python
plt.stem(n,x_n_undersampled )
plt.title("Under sampled by 2")
plt.xlabel("Time(S)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

#DFT Of the Signal

```python
# Sampling the signal with 500 HZ
fs = 500
N = 512
n = np.arange(N) / fs 

X_n = 10*np.sin(2*np.pi *40*n) + 20*np.sin(2*np.pi*80*n) + 40*np.sin(2*np.pi*160*n)

print(len(X_n))
print(X_n)
plt.plot(n,X_n)
plt.title("Sampling Signal")
```

```python
# DFT
def DFT(x):
    N = len(x)
    X = np.zeros(N, dtype=complex)
    for k in range(N):
        for n in range(N):
            X[k] += x[n] * np.exp(-2j * np.pi * k * n / N)
    return X
```

```python
X_dft = DFT(X_n)
magnitude = np.abs(X_dft)

print(X_dft)
plt.stem(n,magnitude)
plt.title("DFT")
plt.xlabel("Time(S)")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()
```

```python
phase = np.angle(X_dft)*180/np.pi
plt.stem(n, phase)
plt.title("Phase of the signal")
plt.xlabel("Time(S)")
plt.ylabel("Phase")
plt.grid(True)
plt.show()
```
