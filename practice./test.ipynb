```python
# write program to generate unit step signal
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(-10, 10, 1)
x = np.zeros(len(n))
x[n >= 0] = 1

plt.stem(n, x)
plt.xlabel('n')
plt.ylabel('x[n]')
plt.title('Unit step signal')
plt.show()
```

```python
#Exponential signal  of x2 = 0.8 ^(n)

import numpy as np

# Define the range of n values
n= np.arange(0, 20)

# Calculate the corresponding x values using the formula x = 0.8^n
x= 0.8 ** n


plt.stem(n, x)
plt.xlabel('n')
plt.ylabel('x[n]')
plt.title('Exponential signal')
plt.show()

```

```python
#generating Ramp signal 
import numpy as np
import matplotlib.pyplot as plt

length = 10

ramp_signal = np.arange(length)

# Plot the ramp signal
plt.stem(ramp_signal)
plt.xlabel('Index')
plt.ylabel('Amplitude')
plt.title('Ramp Signal')
plt.grid(True)
plt.show()
```

```python
# Generating Sine wave signal

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine Wave')
plt.grid(True)
plt.show()
```

```python
# Generating Cosine wave signal

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)
y = np.cos(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine Wave')
plt.grid(True)
plt.show()
```
