```python
# Generating the orginal Signal

import numpy as np
import matplotlib.pyplot as plt

# Define the time range
t = np.linspace(0, 0.02, 500)  
f = 500
# Generate the signal
x = 5 * np.sin(2 * np.pi * f * t + np.radians(90))

# Plot the signal
plt.figure(figsize=(10, 5))
plt.plot(t, x, color='blue')
plt.title('Signal x(t) = 5*sin(2*pi*500*t + 90)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python

# Sampling with a preferred sampling frequency (10 times the signal frequency)

import numpy as np
import matplotlib.pyplot as plt


f = 500  
A = 5  
phi = np.radians(90) 


T = 1 / f  # Period of the signal
t = np.arange(0, 0.02, 1 / (10 * f))  
print(f"The number of Sample is  {len(t)}")

# Generate the signal
x = A * np.sin(2 * np.pi * f * t + phi)

# Plot the signal
plt.figure(figsize=(10, 5))
plt.stem(t, x) # Use the stem plot or we can ue plt.plot(t, x, color='blue'
plt.title('Signal x(t) = 5*sin(2*pi*500*t + 90)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
# Shifting the Signal 

import numpy as np
import matplotlib.pyplot as plt
 
t0 = 0.001  
print("the Number of Sample is", len(t))

plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 1)
plt.stem(t, x)
plt.title('Orginal Signal x(t) = 5*sin(2*pi*500*t + 90)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
print(f"The number of Sample is  {len(t)}")

# Generate the signal with Shotong 
t = t + 1
x = A * np.sin(2 * np.pi * f * t + phi)  

# Plot the signal
plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 2)
plt.stem(t, x)
plt.title('shifting Signal x(t) = 5*sin(2*pi*500*(t - t0) + 90)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
plt.plot(t,x ,label="Orginal Signal")


t_reversed = -t[::-1]
x_reversed = x[::-1]
plt.plot(t_reversed, x_reversed, label = " REversed Signal")
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.title("Orginal Signal VS reversed Signal")
plt.show()
```

```python
#Up Sampling with a factor of 5

import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
f = 500
A = 5
phi = np.radians(90)
t0 = 0.0005

# Define the time range
t_orginal = np.linspace(0, 0.01, 100)
upsample_factor = 2
t_upsampled = np.linspace(0, 0.01, 100 * upsample_factor)

# Generate the signal
x_orginal = A * np.sin(2 * np.pi * f * t_orginal + phi)
x_upsampled = A * np.sin(2 * np.pi * f * t_upsampled + phi)

plt.figure(figsize=(10,5))

print(f"The number of Sample is  {len(t_orginal)}")
plt.subplot(2,1,1)
plt.stem(t_orginal, x_orginal)
plt.title('Orginal Signal x(t) = 5*sin(2*pi*500*t) + *90')
plt.xlabel('Time(s)')
plt.ylabel("Amplitude")
plt.grid(True)

print(f"The number of Up Sample is  {len(t_upsampled)}")
plt.subplot(2,1,2)
plt.stem(t_upsampled, x_upsampled)
plt.title("Upsampled Signal")
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.grid(True)


plt.tight_layout()
plt.show()
```
