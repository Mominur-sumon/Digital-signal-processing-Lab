```python
#Convultion between two sequence

import matplotlib.pyplot as plt 
import numpy as np 


x = input("Enter the first sequece :" )
h = input("Enter the second Sequence :")


# Convert input strings to arrays
x = np.array([int(i) for i in x.split()])
h = np.array([int(i) for i in h.split()])

y = np.convolve(x,h)


plt.subplot(2,2,1)
plt.stem(x)
plt.title("Input Sequence")

plt.subplot(2,2,2)
plt.stem(h)
plt.title("Impulse Sequence")

plt.subplot(2,2,3)
plt.stem(y)
plt.title("Output Sequence")
plt.figure(figsize=(10, 8))
plt.show()
```
