```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Sine Wave 
def sine_wave(Amplitude, phase):
    phase_radian = np.radians(phase)
    x =  np.linspace(start=-100, stop=100, num=100)
    y = Amplitude * np.sin(2*np.pi*x + phase_radian)
    plt.title(f'Sine Wave: A({Amplitude}), P({phase} Degree)')
    plt.grid(True, which='both')
    plt.axhline(color ='r')
    plt.axvline(color ='r')
    plt.plot(x, y)
    plt.show()
```

```python
sine_wave(Amplitude=2, phase=90)
```

```python
# Cosine Wave

def cosine_wave(amplitude, phase):
    phase_radian = np.radians(phase)
    x =  np.linspace(start=-100, stop=100, num=100)
    y = amplitude * np.cos(2*np.pi*x + phase_radian)
    plt.title(f'Cosine Wave: A({amplitude}), P({phase} Degree)')
    plt.grid(True, which='both')
    plt.axhline(color ='r')
    plt.axvline(color ='r')
    plt.plot(x, y)
    plt.show()
```

```python
cosine_wave(amplitude=2, phase=90)
cosine_wave(amplitude=2, phase=0)
```

```python
# Unit Step signal generation
def unit_step_signal(start, end, sampling_rate):
    x = np.linspace(start, end, num= int((end-start)*sampling_rate)+1)
    y= []
    for i in x:
        if i>=0:
            y.append(1)
        else:
            y.append(0)

    plt.title(f'Unit Step Signal: {start} to {end}')
    plt.grid(True, which='both')
    plt.axhline(color ='r')
    plt.axvline(color ='r')
    plt.stem(x, y)
    plt.grid()
    plt.show()            
```

```python
unit_step_signal(-5, 5, 3)
```

```python
def ramp(start, end, sampling_rate):
    x = np.linspace(start= start, stop=end, num = int((end-start)*sampling_rate)+1)
    y = []
    for i in x:
        if(i<0):
            y.append(0)
        else:
            y.append(i)
    plt.title(f'Ramp Signal: {start} to {end}')
    plt.grid(True, which='both')
    plt.axhline(color ='r')
    plt.axvline(color ='r')
    plt.stem(x, y)
    plt.grid()
    plt.show()        

```

```python
ramp(-5, 5, 3)
```

```python
# Exponential Signal

def exponential(start,end, sampling_rate):
    x = np.linspace(start= start, stop=end, num = int((end-start)*sampling_rate)+1)
    y = np.exp(x)

    plt.title(f'Exponential Signal: {start} to {end}')
    plt.grid(True, which='both')
    plt.axhline(color ='r')
    plt.axvline(color ='r')
    plt.stem(x, y)
    plt.grid()
    plt.show()
```

```python
exponential(-5, 5, 3)
```
