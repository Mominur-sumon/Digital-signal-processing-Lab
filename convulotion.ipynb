```python
import matplotlib.pyplot as plt 
import numpy as np 


x = input("Enter the first sequece :" )
h = input("Enter the second Sequence :")


# Convert input strings to arrays
x = np.array([int(i) for i in x.split()])
h = np.array([int(i) for i in h.split()])

print("x[n] = ", x)
print("h[n] = ", h)
```

```python
x_len = len(x)
h_len = len(h)

len_output = x_len + h_len - 1
output = np.zeros(len_output)
    
for i in range(len_output):
    for j in range(h_len):
        if i - j >= 0 and i - j < x_len:
            output[i] += x[i - j] * h[j]
```

```python
output
```

```python
plt.figure(figsize=(10,6))

plt.subplot(2,2,1)
plt.title("x(n) Sequence")
plt.xlabel("n")
plt.ylabel("x(n)")
plt.grid(True)
plt.stem(x)


plt.subplot(2,2,2)
plt.title("h(n) Sequence")
plt.xlabel("n")
plt.ylabel("h(n)")
plt.grid(True)
plt.stem(h)

plt.subplot(2,2,3)
plt.title("Convolution of x(n) and h(n)")
plt.xlabel("n")
plt.ylabel("y(n) = x(n) * h(n)")
plt.grid(True)
plt.stem(output)
```
