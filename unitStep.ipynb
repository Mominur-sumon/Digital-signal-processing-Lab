```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def unit_step(x):
    return 1 * (x > 0)

# Genereate data
# t = np.linspace(-10, 10, 100)
t = np.arange(-5, 5, 0.1)

unit_step_signal = unit_step(t)
```

```python
# ploting the signal 
plt.plot(t, unit_step_signal, label='Unit Step Signal')
plt.title('Unit Step Signal')
plt.xlabel('t')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```
