```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Given Function:
#y = 5 sin(2 * pi * 500t + 90)
```

```python
# Variables
start = 0
stop = 0.01
phase = 90
frequency = 500
amplitude = 5
sample_rate = 4000
sample_count = sample_rate * (stop - start)

# COnversion 
phase_radian = np.radians(phase)

# Base 
t_base = np.linspace(start = start, stop=stop, num= int(sample_count))
y_base = amplitude * np.sin(2 * np.pi * frequency * t_base + phase_radian)
```

```python
# Generate Continuous signal

plt.figure(figsize=(10, 6))
plt.plot(t_base, y_base)
plt.title('Continuous Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

```python
# Generate Discrete signal
plt.figure(figsize=(10, 6))
plt.stem(t_base, y_base)
plt.title('Discrete Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

```python
# Shifting the signal
shifted_t = t_base + 0.5
shifted_y = amplitude * np.sin(2 * np.pi * frequency * shifted_t + phase_radian)

plt.figure(figsize=(10, 6))
# plt.plot(shifted_t, shifted_y)
plt.stem(shifted_t, shifted_y)
plt.title(f'Shifted Signal {shifted_t[0]}')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

```python
# Reverse the signal
reversed_t = -t_base
reversed_y = amplitude * np.sin(2 * np.pi * frequency * reversed_t + phase_radian)
plt.figure(figsize=(10, 6))
plt.stem(reversed_t, reversed_y)
plt.stem(t_base, y_base, 'r')
plt.title('Reversed Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid()

plt.show()
```
